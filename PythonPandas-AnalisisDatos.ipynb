```python
import pandas as pd
```

```python
dfNacion = pd.read_csv(r'.\Dataframe\proyecciones_jurisdicciones_2022_2040_base.csv', delimiter = ';')
#Dataframes con sus relaciones
dfEdad = pd.read_excel(r'.\Dataframe\Edad.xlsx')
dfProvincias = pd.read_excel(r'.\Dataframe\Provincias.xlsx')
dfSexo = pd.read_excel(r'.\Dataframe\Sexo.xlsx')
```

```python
dfNacion.head(3)
```

```python
resultado = dfProvincias.join(dfNacion, lsuffix = '_prov', rsuffix = '_df')
```

```python
resultado = dfProvincias.set_index('ID_Jurisdiccion').join(dfNacion.set_index('ID_Jurisdiccion'))
```

```python
resultado
```

```python
resultado2 = dfSexo.join(resultado, lsuffix = '_sex', rsuffix = '_dfr')
```

```python
resultado2 = dfSexo.set_index('ID_Sexo').join(resultado.set_index('ID_Sexo'))
```

```python
resultado2
```

```python
resultado3 = dfEdad.join(resultado2, lsuffix = '_edad', rsuffix = '_dfe')
```

```python
resultado3 = dfEdad.set_index('ID_Edad').join(resultado2.set_index('ID_Edad'))
```

```python
resultado3
```

```python
resultado_filtrado = resultado3.copy()
```

```python
resultado_filtrado['Poblacion'] = pd.to_numeric(resultado_filtrado['Poblacion'], errors='coerce').fillna(0)
```

```python
resultado_filtrado.head(5)
```

```python
promedio_agregado = resultado_filtrado.groupby(['Provincia', 'Anios', 'Genero'])['Poblacion'].mean().reset_index()
```

```python
promedio_agregado = promedio_agregado.rename(columns={'Poblacion': 'PoblacionPromedio'})
```

```python
promedio_agregado['PoblacionPromedio'] = promedio_agregado['PoblacionPromedio'].round(0).astype(int)
```
